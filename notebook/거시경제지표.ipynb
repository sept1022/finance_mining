```python
import requests 
import json

# print('http://ecos.bok.or.kr/api/StatisticSearch/LO6O7ERYMXF2UM6AS812/json/kr/1/10000/028Y001/MM/198701/202108/BEEA42/BEEA422/?/')

## 정의된 OpenAPI URL을 호출합니다.

def get_data(url):
    response = requests.get(url)


    # ## http 요청이 성공했을때 API의 리턴값을 가져옵니다.

    if response.status_code == 200:

        try:

            contents = response.text
            value = json.loads(contents)
            return value
        except Exception as e:

            print(str(e))
    return None
api_key = 'LO6O7ERYMXF2UM6AS812'
code = '028Y001'
duration = 'MM'
begin = '198701'
end = '202108'

for_three_key = 'BEEA42'
for_ten_key = 'BEEA422'

url = 'http://ecos.bok.or.kr/api/StatisticSearch/{api_key}/json/kr/1/10000/{code}/{duration}/{begin}/{end}/{cond}'

    
three_year = get_data(url.format(api_key=api_key, code=code, duration=duration, begin=begin, end=end, cond=for_three_key))
ten_year = get_data(url.format(api_key=api_key, code=code, duration=duration, begin=begin, end=end, cond=for_ten_key))

table = dict()
for item in ten_year['StatisticSearch']['row']:
    table.setdefault(item['TIME'], dict())
    table[item['TIME']]['ten'] = item['DATA_VALUE']

for item in three_year['StatisticSearch']['row']:
    if item['TIME'] not in table:
        continue
    table[item['TIME']]['three'] = item['DATA_VALUE']

for k, v in table.items():
    v['spread'] = v['ten'] - v['three']
# for l, r in zip(three_year['StatisticSearch']['row'], ten_year['StatisticSearch']['row']):
#     print(l, r)
```
